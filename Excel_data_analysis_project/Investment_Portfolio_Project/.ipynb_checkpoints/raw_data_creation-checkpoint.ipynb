```python
import yfinance as yf
import pandas as pd
```

```python
tcs = yf.download(
    "TCS.NS",                 # ticker as positional argument
    start="2020-01-01",
    end="2024-12-31",
    interval="1mo",
    auto_adjust=False,
    progress=False
)

tcs.reset_index(inplace=True)
tcs.to_csv("TCS_NS_Monthly_2020_2024.csv", index=False)

print("Download completed successfully!")
print(tcs.head())
```

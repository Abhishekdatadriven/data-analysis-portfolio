```python
import pandas as pd
import numpy as np
import random
from datetime import timedelta
```

```python
np.random.seed(42)

users = []

for user_id in range(1, 5001):
    signup_date = pd.to_datetime(
        np.random.choice(pd.date_range("2021-01-01", "2024-12-31"))
    )
    users.append([
        user_id,
        signup_date,
        np.random.choice(["USA", "India", "UK", "Canada"]),
        np.random.choice(["Mobile", "TV", "Tablet"])
    ])

users = pd.DataFrame(
    users,
    columns=["user_id", "signup_date", "country", "device"]
)

```

```python
users.to_csv("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/OTT-Streaming-Analytics/data/raw/users_raw.csv", index=False)
```

```python
content_count = 300

content_ids = range(1, content_count + 1)

genres = ["Drama", "Comedy", "Action", "Thriller", "Romance", "Documentary"]

content = pd.DataFrame({
    "content_id": content_ids,
    "title": [f"Content_{i}" for i in content_ids],
    "genre": np.random.choice(genres, content_count),
    "duration_min": np.random.randint(60, 180, content_count),
    "release_year": np.random.randint(2000, 2025, content_count)
})


```

```python
content.to_csv("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/OTT-Streaming-Analytics/data/raw/content_raw.csv", index=False)
```

```python
streams = []

for _, row in users.iterrows():
    user_id = row["user_id"]
    signup_date = row["signup_date"]

    sessions = np.random.randint(5, 40)

    for _ in range(sessions):
        watch_date = signup_date + pd.to_timedelta(
            np.random.randint(0, 365), unit="D"
        )

        streams.append([
            watch_date,
            user_id,
            np.random.randint(1, 300),   # content_id
            np.random.randint(5, 120)    # watch_minutes
        ])

fact_streaming = pd.DataFrame(
    streams,
    columns=["date", "user_id", "content_id", "watch_minutes"]
)


```

```python
fact_streaming.to_csv("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/OTT-Streaming-Analytics/data/raw/viewing_activity_raw.csv", index=False)
```

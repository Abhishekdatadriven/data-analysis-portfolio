```python
import pandas as pd
```

```python
users = pd.read_csv("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/OTT-Streaming-Analytics/data/processed/clean/users_clean.csv")
content = pd.read_csv("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/OTT-Streaming-Analytics/data/processed/clean/content_clean.csv")
viewing = pd.read_csv("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/OTT-Streaming-Analytics/data/processed/clean/viewing_activity_clean.csv")
```

```python
dim_user = users[[
    "user_id",
    "signup_date",
    "country",
    "device"
]].copy()
```

```python
dim_user.head()
```

```python
#Create DIM_CONTENT
dim_content = content[[
    "content_id",
    "title",
    "genre",
    "release_year",
    "duration_min"
]].copy()

```

```python
dim_content.head()
```

```python
#Create DIM_DATE
date_range = pd.date_range(
    start=viewing["date"].min(),
    end=viewing["date"].max()
)

dim_date = pd.DataFrame({"date": date_range})
```

```python
dim_date["year"] = dim_date["date"].dt.year
dim_date["month"] = dim_date["date"].dt.month
dim_date["month_name"] = dim_date["date"].dt.strftime("%b")
dim_date["quarter"] = dim_date["date"].dt.to_period("Q").astype(str)
dim_date["day"] = dim_date["date"].dt.day
```

```python
dim_date.head()
```

```python
#Create FACT_VIEWING
fact_viewing = viewing[[
    "date",
    "user_id",
    "content_id",
    "watch_minutes",
    "completed_flag"
]].copy()
```

```python
fact_viewing["watch_sessions"] = 1
```

```python
fact_viewing.head()
```

```python
#Validation
assert fact_viewing["user_id"].isin(dim_user["user_id"]).all()
assert fact_viewing["content_id"].isin(dim_content["content_id"]).all()
```

```python
dim_user.to_csv("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/OTT-Streaming-Analytics/data/processed/modeled/dim_user.csv", index=False)
dim_content.to_csv("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/OTT-Streaming-Analytics/data/processed/modeled/dim_content.csv", index=False)
dim_date.to_csv("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/OTT-Streaming-Analytics/data/processed/modeled/dim_date.csv", index=False)
fact_viewing.to_csv("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/OTT-Streaming-Analytics/data/processed/modeled/fact_viewing.csv", index=False)
```

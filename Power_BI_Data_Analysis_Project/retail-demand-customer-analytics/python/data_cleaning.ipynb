```python
import pandas as pd
```

```python
sales = pd.read_csv("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/retail-demand-customer-analytics/data/raw/sales_raw.csv", parse_dates=["date"])
customers = pd.read_csv("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/retail-demand-customer-analytics/data/raw/customers_raw.csv")
products = pd.read_csv("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/retail-demand-customer-analytics/data/raw/products_raw.csv")
stores = pd.read_csv("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/retail-demand-customer-analytics/data/raw/stores_raw.csv")
```

```python
# Remove duplicates
sales.drop_duplicates(inplace=True)
```

```python
# Fix data types
sales["quantity"] = sales["quantity"].astype(int)
sales["revenue"] = sales["revenue"].astype(float)
```

```python
# Filter invalid records
sales = sales[
    (sales["quantity"] > 0) &
    (sales["unit_price"] > 0) &
    (sales["discount"] >= 0) &
    (sales["discount"] <= 0.5)
]
```

```python
sales.head()
```

```python
customers = customers.drop_duplicates(subset="customer_id")
```

```python
customers.head()
```

```python
products = products.drop_duplicates(subset="product_id")
```

```python
products.head()
```

```python
stores = stores.drop_duplicates(subset="store_id")
```

```python
stores["store_type"] = stores["store_type"].fillna("Unknown")
```

```python
print("Sales rows:", sales.shape)
print("Customers rows:", customers.shape)
print("Products rows:", products.shape)
print("Stores rows:", stores.shape)
```

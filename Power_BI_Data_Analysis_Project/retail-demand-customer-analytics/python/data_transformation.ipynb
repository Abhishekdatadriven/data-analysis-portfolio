```python
import pandas as pd
```

```python
sales = pd.read_csv("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/retail-demand-customer-analytics/data/processed/clean/sales_clean.csv", parse_dates=["date"])
customers = pd.read_csv("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/retail-demand-customer-analytics/data/processed/clean/customers_clean.csv")
products = pd.read_csv("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/retail-demand-customer-analytics/data/processed/clean/products_clean.csv")
stores = pd.read_csv("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/retail-demand-customer-analytics/data/processed/clean/stores_clean.csv")
```

```python
# -----------------------------
# Date Dimension
# -----------------------------
dim_date = pd.DataFrame({
    "date": pd.to_datetime(sales["date"].unique())
})

dim_date["year"] = dim_date["date"].dt.year
dim_date["month"] = dim_date["date"].dt.month
dim_date["month_name"] = dim_date["date"].dt.strftime("%b")
dim_date["year_month"] = dim_date["date"].dt.to_period("M").astype(str)
```

```python
dim_date.head()
```

```python
#Create CUSTOMER DIMENSION
dim_customer = customers.copy()
```

```python
dim_customer.head()
```

```python
#Create PRODUCT DIMENSION
dim_product = products.copy()
```

```python
dim_product.head()
```

```python
#Create STORE DIMENSION
dim_store = stores.copy()
```

```python
dim_store.head()
```

```python
# -----------------------------
# Fact Table
# -----------------------------
fact_sales = sales.merge(dim_date, on="date", how="left")
```

```python
fact_sales.head()
```

```python
print("Fact rows:", fact_sales.shape)
print("Dim Date rows:", dim_date.shape)
print("Dim Customer rows:", dim_customer.shape)
print("Dim Product rows:", dim_product.shape)
print("Dim Store rows:", dim_store.shape)
```

```python
assert fact_sales["customer_id"].isin(dim_customer["customer_id"]).all()
assert fact_sales["product_id"].isin(dim_product["product_id"]).all()
assert fact_sales["store_id"].isin(dim_store["store_id"]).all()
```

```python
# -----------------------------
# SAVE MODELED DATA
# -----------------------------
fact_sales.to_csv("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/retail-demand-customer-analytics/data/processed/modeled/fact_sales.csv", index=False)
dim_date.to_csv("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/retail-demand-customer-analytics/data/processed/modeled/dim_date.csv", index=False)
dim_customer.to_csv("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/retail-demand-customer-analytics/data/processed/modeled/dim_customer.csv", index=False)
dim_product.to_csv("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/retail-demand-customer-analytics/data/processed/modeled/dim_product.csv", index=False)
dim_store.to_csv("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/retail-demand-customer-analytics/data/processed/modeled/dim_store.csv", index=False)
```

```python
import pandas as pd
from pathlib import Path
```

```python
raw_path = Path("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/Investment-Performance-Risk-Analyzer/data/raw")

files = {
    "TATAMOTORS": "tatamotors_prices.csv",
    "ICICIBANK": "icicibank_prices.csv",
    "ITC": "itc_prices.csv",
    "NIFTY50": "nifty50_prices.csv",
    "GOLD_ETF": "gold_etf_prices.csv",
    "DEBT_FUND": "debt_fund_prices.csv"
}
```

```python
all_data = []

for asset, file in files.items():
    df = pd.read_csv(raw_path / file)

    # Standardize column names
    df = df.rename(columns={"Date": "date", "Close": "price"})
    df["date"] = pd.to_datetime(df["date"])

    # Sort by date 
    df = df.sort_values("date")

    # Add asset name
    df["asset"] = asset

    # Calculate Monthly Return
    df["monthly_return"] = df["price"].pct_change()

    all_data.append(df)
```

```python
consolidated_df = pd.concat(all_data, ignore_index=True)
consolidated_df.head()
```

```python
consolidated_df = consolidated_df.dropna()
consolidated_df.head()
```

```python
consolidated_df = consolidated_df.reset_index(drop=True)
consolidated_df.head()
```

```python
output_path = Path("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/Investment-Performance-Risk-Analyzer/data/processed/consolidated_prices.csv")
consolidated_df.to_csv(output_path, index=False)

print("Consolidated dataset created successfully!")
```

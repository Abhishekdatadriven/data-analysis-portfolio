```python
import pandas as pd
import numpy as np
from pathlib import Path
import yfinance as yf
```

```python
tickers = {
    "TATAMOTORS": "TATAMOTORS.NS",
    "ICICIBANK": "ICICIBANK.NS",
    "ITC": "ITC.NS",
    "NIFTY50": "^NSEI",
    "GOLD_ETF": "GOLDBEES.NS",
    "DEBT_FUND": "EBBETF0431.NS"
}

raw_path = Path("C:/Users/Abhi/Desktop/Power_BI_Data_Analysis_Project/Investment-Performance-Risk-Analyzer/data/raw")
raw_path.mkdir(parents=True, exist_ok=True)

for asset, ticker in tickers.items():
    df = yf.download(
        ticker,
        start="2020-01-01",
        end="2024-12-31",
        interval="1mo",
        auto_adjust=True
    )

    df = df.reset_index()[["Date", "Close"]]
    df.to_csv(raw_path / f"{asset.lower()}_prices.csv", index=False)

    print(f"Downloaded {asset}")
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
   "C:/Users/Abhi/Desktop/Python_Data_Analytics_Projects/demand_supply_mismatch_project/data/processed/mismatch_detected_data.csv",
    parse_dates=["date"]
)
df = df.sort_values(["product_id", "date"])
df.head()

    
```

```python
df["lost_sales_qty"] = (
    df["avg_demand_7"] - df["inventory_qty"]
).where(
    df["mismatch_type"] == "Stockout Risk",
    0
).clip(lower=0)
df.head()
```

```python
df["lost_revenue"] = (
    df["lost_sales_qty"] * df["unit_price"]
)
df.head()
```

```python
df["excess_inventory_qty"] = (
    df["inventory_qty"] - df["avg_demand_30"]
).where(
    df["mismatch_type"] == "Overstock Risk",
    0
).clip(lower=0)
df.head()
```

```python
HOLDING_COST_RATE = 0.02
```

```python
df["overstock_cost"] = (
    df["excess_inventory_qty"] *
    df["unit_price"] *
    HOLDING_COST_RATE
)
df.head()
```

```python
impact_summary = pd.DataFrame({
    "Metric": [
        "Total Products",
        "Stockout Risk Days",
        "Overstock Risk Days",
        "Estimated Lost Revenue",
        "Estimated Overstock Cost"
    ],
    "Value": [
        df["product_id"].nunique(),
        (df["mismatch_type"] == "Stockout Risk").sum(),
        (df["mismatch_type"] == "Overstock Risk").sum(),
        round(df["lost_revenue"].sum(), 2),
        round(df["overstock_cost"].sum(), 2)
    ]
})

impact_summary
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
   "C:/Users/Abhi/Desktop/Python_Data_Analytics_Projects/demand_supply_mismatch_project/data/processed/feature_engineered_data.csv",
    parse_dates=["date"]
)

df = df.sort_values(["product_id", "date"])
df.head()
```

```python
df["stockout_risk"] = np.where(
    (df["inventory_coverage"] < 5) &
    (df["demand_change"] > 0),
    1,
    0
)
df.head()
```

```python
df["overstock_risk"] = np.where(
    (df["inventory_coverage"] > 30) &
    (df["demand_change"] > 0),
    1,
    0
)
df.head()
```

```python
df["mismatch_type"] = np.select(
    [
        df["stockout_risk"] == 1,
        df["overstock_risk"] == 1
    ],
    [
        "Stockout Risk",
        "Overstock Risk"
    ],
    default="Balanced"
)
df.head()
```

```python
df["risk_score"] = (
    df["inventory_stress"] * 0.5 +
    df["demand_volatility"].fillna(0) * 0.3 +
    abs(df["demand_change"]) * 0.2
)
df.head()
```

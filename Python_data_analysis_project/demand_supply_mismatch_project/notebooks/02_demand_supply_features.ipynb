```python
import pandas as pd
import numpy as np
```

```python
# Load cleaned data
df = pd.read_csv("C:/Users/Abhi/Desktop/Python_Data_Analytics_Projects/demand_supply_mismatch_project/data/raw/demand_supply_raw.csv", parse_dates=["date"])

df = df.sort_values(["product_id", "date"])
df.head()
```

```python
df["avg_demand_7"] = (
    df.groupby("product_id")["sales_qty"]
      .transform(lambda x: x.rolling(7, min_periods=1).mean())
)
df.head()
```

```python
df["avg_demand_30"] = (
    df.groupby("product_id")["sales_qty"]
      .transform(lambda x: x.rolling(30, min_periods=1).mean())
)
df.head()
```

```python
df["demand_volatility"] = (
    df.groupby("product_id")["sales_qty"]
      .transform(lambda x: x.rolling(7, min_periods=1).std())
)
df.head()
```

```python
df["inventory_coverage"] = (
    df["inventory_qty"] / df["avg_demand_7"]
).replace([np.inf, -np.inf], 0).fillna(0)
df.head()
```

```python
df["demand_change"] = (
    df.groupby("product_id")["avg_demand_7"]
      .transform(lambda x: x.diff())
)
df.head()
```

```python
df["inventory_stress"] = (
    (df["avg_demand_7"] / df["inventory_qty"])
).replace([np.inf, -np.inf], 0).fillna(0)
df.head()
```

```python
df.to_csv(
    "C:/Users/Abhi/Desktop/Python_Data_Analytics_Projects/demand_supply_mismatch_project/data/processed/feature_engineered_data.csv",
    index=False
)

df.head()
```

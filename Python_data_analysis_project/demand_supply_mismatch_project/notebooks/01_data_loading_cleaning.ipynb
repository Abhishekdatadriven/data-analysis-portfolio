```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
```

```python
products = ["P001", "P002", "P003", "P004", "P005"]
dates = pd.date_range(start="2023-01-01", end="2024-12-31", freq="D")

data = []

for product in products:
    base_demand = np.random.randint(15, 40)
    inventory = np.random.randint(300, 700)
    price = np.random.randint(50, 200)

    for date in dates:
        demand = np.random.poisson(base_demand)

        # Sales
        sales = min(demand, inventory)
        inventory -= sales

        # ðŸ”§ FIX: Explicit replenishment
        purchase_qty = 0
        if inventory < base_demand * 5:
            purchase_qty = np.random.randint(200, 500)
            inventory += purchase_qty

        data.append([
            date,
            product,
            sales,          # demand
            inventory,
            purchase_qty,   # supply (NOW NON-ZERO)
            price
        ])

df = pd.DataFrame(
    data,
    columns=[
        "date",
        "product_id",
        "sales_qty",
        "inventory_qty",
        "purchase_qty",
        "unit_price"
    ]
)

df.head()
```

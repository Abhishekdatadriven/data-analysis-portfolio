```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("C:/Users/Abhi/Desktop/Python_Data_Analytics_Projects/behavioral-bias-detector/data/processed/TCS_cleaned.csv")
df["date"]=pd.to_datetime(df["date"])
df= df.sort_values("date").reset_index(drop=True)
df.head()
```

```python
df["daily_return"]=df["close"].pct_change()
df.head()
```

```python
df["return_5d"] = df["close"].pct_change(5)
df["return_10d"] = df["close"].pct_change(10)
df.head(20)
```

```python
df["volatility_10d"] = df["daily_return"].rolling(10).std()
df.head(20)
```

```python
df["rolling_max"] = df["close"].cummax()
df["drawdown"] = (df["close"] - df["rolling_max"]) / df["rolling_max"]
df.head(50)
```

```python
df["volume_mean_20d"] = df["volume"].rolling(20).mean()
df["volume_std_20d"] = df["volume"].rolling(20).std()
df["volume_zscore"]=(
    (df["volume"] - df["volume_mean_20d"])/ df["volume_std_20d"]
)
df.head(50)
```

```python
df["high_52w"] = df["close"].rolling(252).max()
df["dist_from_52w_high"] = (df["close"] - df["high_52w"]) / df["high_52w"]
df.head(50)
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.to_csv("C:/Users/Abhi/Desktop/Python_Data_Analytics_Projects/behavioral-bias-detector/data/processed/TCS_features.csv", index=False)
```

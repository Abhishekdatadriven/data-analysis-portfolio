```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
     "C:/Users/Abhi/Desktop/Python_Data_Analytics_Projects/car_insurance_pricing_analytics/data/processed/car_insurance_policy_cleaned.csv",
    parse_dates=["policy_start_date"]
)

df.head()
```

```python
df["premium_per_vehicle_age"] =(
    df["final_premium"] /(df["vehicle_age"] + 1)
)
df.head()
```

```python
df["discount_applied"]  = np.where(
    df["discount_pct"] > 0, 1, 0
)   
```

```python
df["discount_bucket"] = pd.cut(
    df["discount_pct"],
    bins=[-1, 0, 5, 10, 15],
    labels=["No Discount", "Low", "Medium", "High"]
)
df.head()
```

```python
df["avg_vehicle_premium"] = (
    df.groupby("vehicle_type")["final_premium"]
      .transform("mean")
)
df.head()
```

```python
df["premium_deviation_pct"] = (
    (df["final_premium"] - df["avg_vehicle_premium"])
    / df["avg_vehicle_premium"]
) * 100
df.head()
```

```python
df["risk_adjusted_discount"] = (
    df["discount_pct"] / (df["vehicle_age"] + 1)
)
df.head()
```

```python
df["pricing_status"] = np.where(
    df["underwriting_margin"] < 0,
    "Underpriced",
    "Profitable"
)
df.head()
```

```python
df["effective_price_index"] = (
    df["final_premium"] / df["avg_vehicle_premium"]
)
df.head()
```

```python
df.to_csv(
    "C:/Users/Abhi/Desktop/Python_Data_Analytics_Projects/car_insurance_pricing_analytics/data/processed/car_insurance_pricing_features.csv",
    index=False
)
```
